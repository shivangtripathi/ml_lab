```python
#Q9
def caught_challan(a,b):
  if (b==True):
     if (a <= 65):
       return "No challan"
     elif (a>65 and a<=85):
       return "Small Challan"
     elif (a>85):
       return "Heavy Challan"


  else:
    if (a <= 60):
       return "No challan"
    elif (a>60 and a<=80):
         return "Small Challan"
    elif (a>80):
       return "Heavy Challan"


ss5=caught_challan(63,False)       
print(ss5)

ss6=caught_challan(63,True)       
print(ss6)

```
