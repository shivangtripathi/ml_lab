```python
#Q1
import numpy as np
arr=np.empty([10],dtype=int)
arr.fill(5)
print(arr)
print("\n")

#Q2
arr=np.arange(10,51)
print(arr)
print("\n")

#Q3
arr=np.arange(10,51,2,dtype=int)
print(arr)
print("\n")

#Q4
arr=np.arange(0,9).reshape(3,3)
print(arr)
print("\n")

#Q5
arr=np.random.rand(25)
print(arr)
print("\n")

#Q6
arr=np.arange(0,1.01,0.01)
print(arr)
print("\n")

#Q7
arr=np.linspace(0,1,20)
print(arr)
print("\n")

#Q8
arr1=np.arange(1,26).reshape(5,5)
print(arr1)
mat=arr1[2: , 1:]
print("\n\n")
print(mat)

arr2=arr1[0:3,1:2]
print("\n\n")
print(arr2)

print("\n")
print("Sum of all elements is")
total_sum=np.sum(mat)
print(total_sum)
print("\n")
print("Sum of all columns ")
sum_of_row=np.sum(mat,axis=0)
print(sum_of_row)
print("\n")

print("Sum of all rows is")
sum_of_columns=np.sum(mat,axis=1)
print(sum_of_columns)
```
